```python
#Accept string from user
str=input("Enter a string: ")

#Assign vowels to a variable
vowels="aeiouAEIOU"

#Count the number of vowels in given str
i=0
for char in str:
  if char in vowels:
    i+=1

#Display the count
print("Number of vowels in given string= ",i)
```

```python
#Create a function to find factorial
def fact(n):
  if n<0:
    print("Invalid input")
    return None
  elif n==0:
    return 1 #Known value
  else:
    return n*fact(n-1)
  print("Factorial of given integer is ",fact(n))

#Accept an integer from user
n=int(input("Enter a positive integer: "))
fact(n)



```

```python
#Create a list of subjects
subjects=["maths","english","chemistry","physics","biology"]

#Create an empty list for mark
mark=[]

#Accept entered marks into the empty list
for i in range(5):
  mark.append(int(input(f'Enter marks in {subjects[i]} out of 20: ')))

 #Calculate total marks and display it
total_marks=sum(mark)
print("Total marks = ",total_marks,"/100")

#Determine the grade and display it
if total_marks>=90:
  print("Grade = Excellent")
elif total_marks>=65:
  print("Grade = Good")
elif total_marks>=50:
  print("Grade = Average")
elif total_marks>=40:
  print("Grade = Pass")
else:
  print("Grade = Fail")
```

```python
#Accept a number from user
n=int(input("Enter a number: "))

print("\n----Multiplication Table----\n")

#Display Multiplication Table
for i in range(1,11):
  print(f'{n}*{i}={n*i}')
```

```python
#Condition
has_upper=False
has_lower=False
has_digit=False
has_special=False

print("CREATE PASSSWORD")

#Accept password from user
password=input("Enter your password: ")

#Check if the entered password is valid
if len(password)>=8:
  for char in password:
    if char.isupper():
      has_upper=True
    elif char.islower():
      has_lower=True
    elif char.isdigit():
      has_digit=True
    else:
      has_special=True

  # After iterating through all characters, check the conditions
  if not has_lower:
    print("Password must contain at least one lower case letter")
  elif not has_upper:
    print("Password must contain at least one upper case letter")
  elif not has_digit:
    print("Password must contain at least one digit")
  elif not has_special:
    print("Password must contain at least one special character")
  else:
    print("Password created successfully")
else:
  print("Password must contain at least 8 characters!")
```
